```python
import pandas as pd
import numpy as np 
from scipy import stats
import matplotlib.pyplot as plt
```

```python
filelocation = 'D:\\ITS\\semester 1\\Statprob\\Tugas\\Dataset.xlsx'
df = pd.read_excel(filelocation)
```

```python
pivot_total = pd.pivot_table(df, values=['Penjualan A (pcs)', 'Penjualan B (pcs)'], index='Tanggal', aggfunc='sum')
print(pivot_total)
```

```python
penjualan_a = df['Penjualan A (pcs)']
penjualan_b = df['Penjualan B (pcs)']

# Mean
mean_a = np.floor(penjualan_a.mean())
mean_b = np.floor(penjualan_b.mean())

# Median
median_a = np.floor(penjualan_a.median())
median_b = np.floor(penjualan_b.median())

# Modus
mode_a = np.floor(penjualan_a.mode()[0])  
mode_b = np.floor(penjualan_b.mode()[0])  

# Range
range_a = np.floor(penjualan_a.max() - penjualan_a.min())
range_b = np.floor(penjualan_b.max() - penjualan_b.min())

# IQR 
iqr_a = np.floor(stats.iqr(penjualan_a))
iqr_b = np.floor(stats.iqr(penjualan_b))

# Varian
var_a = np.floor(penjualan_a.var())
var_b = np.floor(penjualan_b.var())

# Standar Deviasi
std_dev_a = np.floor(penjualan_a.std())
std_dev_b = np.floor(penjualan_b.std())

# Skewness 
skewness_a = np.floor(penjualan_a.skew())
skewness_b = np.floor(penjualan_b.skew())

# Kurtosis
kurtosis_a = np.floor(penjualan_a.kurtosis())
kurtosis_b = np.floor(penjualan_b.kurtosis())

#Q1 & Q2
Q1a = df['Penjualan A (pcs)'].quantile(0.25)
Q3a = df['Penjualan A (pcs)'].quantile(0.75)
Q1b = df['Penjualan B (pcs)'].quantile(0.25)
Q3b = df['Penjualan B (pcs)'].quantile(0.75)

# Menampilkan hasil
print("Analisis Penjualan A:")
print(f"Mean: {mean_a}")
print(f"Median: {median_a}")
print(f"Modus: {mode_a}")
print(f"Range: {range_a}")
print(f"IQR: {iqr_a}")
print(f"Varian: {var_a}")
print(f"Standar Deviasi: {std_dev_a}")
print(f"Skewness: {skewness_a}")
print(f"Kurtosis: {kurtosis_a}")
print(f"Q1: {Q1a}")
print(f"Q3: {Q3a}")

print("\nAnalisis Penjualan B:")
print(f"Mean: {mean_b}")
print(f"Median: {median_b}")
print(f"Modus: {mode_b}")
print(f"Range: {range_b}")
print(f"IQR: {iqr_b}")
print(f"Varian: {var_b}")
print(f"Standar Deviasi: {std_dev_b}")
print(f"Skewness: {skewness_b}")
print(f"Kurtosis: {kurtosis_b}")
print(f"Q1: {Q1b}")
print(f"Q3: {Q3b}")
```

```python

bins_a = [0, 50, 100, 200, 300, 400, 500]  
labels_a = [f"{bins_a[i]}-{bins_a[i+1]}" for i in range(len(bins_a)-1)]
df['Interval A'] = pd.cut(df['Penjualan A (pcs)'], bins=bins_a, labels=labels_a, right=False)


freq_a = df['Interval A'].value_counts().sort_index()


print("Tabel Distribusi Frekuensi Penjualan A:")
print(freq_a)
```

```python
# Histogram A
plt.figure(figsize=(10, 5))
plt.hist(df['Penjualan A (pcs)'], bins=bins_a, edgecolor='black')
plt.title('Histogram Penjualan A')
plt.xlabel('Jumlah Penjualan A (pcs)')
plt.ylabel('Frekuensi')

plt.show()
```

```python

mid_points_a = [(bins_a[i] + bins_a[i + 1]) / 2 for i in range(len(bins_a) - 1)]
freq_values_a = freq_a.values

plt.figure(figsize=(10, 5))
plt.plot(mid_points_a, freq_values_a, marker='o', linestyle='-', color='b')
plt.title('Poligon Frekuensi Penjualan A')
plt.xlabel('Jumlah Penjualan A (pcs)')
plt.ylabel('Frekuensi')
plt.grid()
plt.xticks(mid_points_a, labels=labels_a)
plt.show()
```

```python
# Boxplot A
plt.figure(figsize=(10, 5))
plt.boxplot(df['Penjualan A (pcs)'], vert=False, patch_artist=True)
plt.title('Boxplot Penjualan A')
plt.xlabel('Jumlah Penjualan A (pcs)')
plt.show()
```

```python

bins_b = [0, 50, 100, 200, 300, 400, 500]  
labels_b = [f"{bins_b[i]}-{bins_b[i+1]}" for i in range(len(bins_b)-1)]
df['Interval B'] = pd.cut(df['Penjualan B (pcs)'], bins=bins_b, labels=labels_b, right=False)


freq_b = df['Interval B'].value_counts().sort_index()


print("Tabel Distribusi Frekuensi Penjualan B:")
print(freq_b)
```

```python
# Histogram B
plt.figure(figsize=(10, 5))
plt.hist(df['Penjualan B (pcs)'], bins=bins_b, edgecolor='black')
plt.title('Histogram Penjualan B')
plt.xlabel('Jumlah Penjualan B (pcs)')
plt.ylabel('Frekuensi')

plt.show()
```

```python
#Polygon B
mid_points_b = [(bins_b[i] + bins_b[i + 1]) / 2 for i in range(len(bins_b) - 1)]
freq_values_b = freq_b.values

plt.figure(figsize=(10, 5))
plt.plot(mid_points_b, freq_values_b, marker='o', linestyle='-', color='b')
plt.title('Poligon Frekuensi Penjualan B')
plt.xlabel('Jumlah Penjualan B (pcs)')
plt.ylabel('Frekuensi')
plt.grid()
plt.xticks(mid_points_b, labels=labels_b)
plt.show()
```
